### Dataset Swagger

```python
# --------------------------------------------------------------------
# Chargement de Dataset Swagger
# --------------------------------------------------------------------

import requests
import json 
import pandas as pd, numpy as np

# Chargement des données Swagger dans JSON
SWAGGER = "http://127.0.0.1:8000/openapi.json"
response = requests.get(SWAGGER)
data = response.json()

with open("swagger.json", "w") as f:
    json.dump(data, f, indent=4)  
df = pd.json_normalize(data)
```
